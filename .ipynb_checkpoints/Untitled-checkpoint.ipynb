```python
import pandas as pd

# Loading the dataset 
df = pd.read_csv(r"C:\IRONHACK\WEEK_2\DAY_5_LAB_2\amz_uk_price_prediction_dataset.csv")

# Inspect the data
df.head()
```

```python
# Creating crosstab
category_bestseller_ct = pd.crosstab(df['category'], df['isBestSeller'], margins=True)

# rename columns for clarity
category_bestseller_ct.columns = ['Not BestSeller', 'BestSeller', 'Total']

# Displaying crosstab
print(category_bestseller_ct)
```

```python
# Grouping by category and calculate proportion of best-sellers
bestseller_proportion = df.groupby('category')['isBestSeller'].mean() * 100  # multiply by 100 to get percentage

# Sorting categories by best-seller proportion in descending order
bestseller_proportion_sorted = bestseller_proportion.sort_values(ascending=False)

# Displaying the top categories
print(bestseller_proportion_sorted.head(10))
```

```python
from scipy.stats import chi2_contingency

# Chi-square test
chi2, p, dof, expected = chi2_contingency(category_bestseller_ct)

print(f"Chi-square statistic: {chi2:.2f}")
print(f"Degrees of freedom: {dof}")
print(f"P-value: {p:.5f}")
```

```python
import numpy as np
# Calculating Cramér's V
n = category_bestseller_ct.sum().sum()  # total observations
cramers_v = np.sqrt(chi2 / (n * (min(category_bestseller_ct.shape)-1)))

print(f"Cramér's V: {cramers_v:.3f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_theme(style="whitegrid", font_scale=1.2)

# Detect columns
cols = category_bestseller_ct.columns
bestseller_col = "BestSeller"
other_col = [c for c in cols if c != bestseller_col][0]

# Select top 10 categories
top_categories = (
    category_bestseller_ct
    .sort_values(by=bestseller_col, ascending=False)
    .head(10)
)

# Modern color palette
colors = ["#d9d9d9", "#4a90e2"]

fig, ax = plt.subplots(figsize=(12, 7))

top_categories.plot(
    kind="barh",
    stacked=True,
    color=colors,
    edgecolor="black",
    linewidth=0.5,
    ax=ax
)

ax.set_title(
    "Top 10 Product Categories by Best-Seller Count",
    fontsize=18,
    weight="bold",
    pad=15
)
ax.set_xlabel("Number of Products")
ax.set_ylabel("Product Category")

# VALUE LABELS
for i, (other, best) in enumerate(
    zip(top_categories[other_col], top_categories[bestseller_col])
):
    ax.text(
        other + best / 2,
        i,
        str(best),
        ha="center",
        va="center",
        color="white",
        fontsize=11,
        weight="bold"
    )

ax.legend([other_col, bestseller_col], title="Product Status")
sns.despine(left=True, bottom=True)

plt.tight_layout()
plt.show()
```

## Exploring Product Prices

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid", palette="Set2", font_scale=1.2)

#----REMOVE PRICE OUTLIERS (IQR METHOD) ----

Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_clean = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)].copy()

print(f"Original rows: {len(df)}")
print(f"Rows after removing price outliers: {len(df_clean)}")

```

```python
# ---- TOP 20 CATEGORIES BY COUNT ----
top20_categories = (
    df_clean['category']
    .value_counts()
    .head(20)
    .index
)

df_top20 = df_clean[df_clean['category'].isin(top20_categories)]


```

```python
# ----  VIOLIN PLOT — TOP 20 CATEGORIES ----

plt.figure(figsize=(14, 8))
sns.violinplot(
    data=df_top20,
    x="price",
    y="category",
    cut=0,
    inner="quartile",
    linewidth=1.2
)

plt.title("Price Distribution Across Top 20 Product Categories", fontsize=16, weight="bold")
plt.xlabel("Price (£)")
plt.ylabel("Product Category")
plt.tight_layout()
plt.show()


```

```python
# ----  VIOLIN PLOT — ALL CATEGORIES (FOR MEDIAN ANALYSIS) ----
plt.figure(figsize=(14, 22))
sns.violinplot(
    data=df_clean,
    x="price",
    y="category",
    cut=0,
    inner="quartile",
    linewidth=1.0
)

plt.title("Price Distribution Across All Product Categories", fontsize=18, weight="bold")
plt.xlabel("Price (£)")
plt.ylabel("Product Category")
plt.tight_layout()
plt.show()


```

```python
# ----  FIND CATEGORY WITH THE HIGHEST MEDIAN PRICE ----
median_prices = df_clean.groupby("category")["price"].median().sort_values(ascending=False)

print("\nTop 10 categories by median price:")
print(median_prices.head(10))
```

# Average Price for the Top 20

```python
# ----  Top 10 categories by product count ----
top10 = df_clean['category'].value_counts().head(10).index
df_top10 = df_clean[df_clean['category'].isin(top10)]

# Average prices
avg_price_top10 = (
    df_top10.groupby("category")["price"]
    .mean()
    .sort_values(ascending=False)
)

```

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    x=avg_price_top10.values,
    y=avg_price_top10.index,
    hue=avg_price_top10.index,   # required in seaborn >= 0.13
    legend=False,                # hide unnecessary legend
    palette="viridis"
)

plt.title("Average Price of Products in Top 10 Categories", fontsize=16, weight="bold")
plt.xlabel("Average Price (£)")
plt.ylabel("Product Category")
plt.tight_layout()
plt.show()
```

# Box Plot showing the distribution of product ratings based on their category

```python
import seaborn as sns
import matplotlib.pyplot as plt

# -------------------------
# Top 10 categories by count
# -------------------------
top10_categories = df['category'].value_counts().head(10).index
df_top10 = df[df['category'].isin(top10_categories)]

plt.figure(figsize=(14, 7))

sns.boxplot(
    data=df_top10,
    x="category",
    y="stars",
    hue="category",         
    legend=False,           
    palette="Set2"
)

plt.title("Distribution of Product Ratings for Top 10 Categories", fontsize=16, weight="bold")
plt.xlabel("Product Category")
plt.ylabel("Customer Rating")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

## Correlation Between Product Price and Rating

```python
# Compute correlation between price and stars
correlation_price_rating = df[['price', 'stars']].corr().iloc[0, 1]

print(f"Correlation between Price and Rating: {correlation_price_rating:.4f}")
```

The correlation between product price and customer rating is –0.125, which is very close to zero. This means there is no meaningful relationship between how much a product costs and how well customers rate it. In practical terms, expensive products are not consistently rated higher or lower than cheaper ones. Customer satisfaction expressed through star ratings, appears to depend more on factors like product quality, expectations, and brand reputation rather than price itself. Overall, price does not significantly influence customer ratings in this dataset.

## Scatter plot (Price vs. Rating)

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='stars', y='price')
plt.title('Scatter Plot of Price vs. Rating')
plt.xlabel('Product Rating (Stars)')
plt.ylabel('Price')
plt.yscale('log')  
plt.show()
```

The scatter plot shows that product prices vary dramatically, ranging from very cheap to extremely expensive items. Most products receive ratings between 2 and 5 stars, with very few at the low end. There isn’t a clear relationship between price and rating—highly rated products can be either inexpensive or costly. A few extreme outliers stand out in the high-price range, but the majority of items are clustered at lower prices. Overall, the plot suggests that paying more doesn’t necessarily guarantee a higher rating, and most products fall into an affordable range regardless of their rating.

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

plt.figure(figsize=(6, 6))
stats.probplot(df['price'], dist="norm", plot=plt)
plt.title('QQ Plot of Product Prices')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.show()
```

The QQ plot indicates that product prices are heavily right-skewed. Most of the points corresponding to lower-priced products lie close to the diagonal line, suggesting that these typical prices roughly follow a normal distribution. However, the points at the high end deviate sharply above the line, highlighting the presence of a few extremely expensive outliers. Overall, this shows that the distribution of prices is not normal, with most products being affordable while a small number are priced very high. Applying a log transformation to the prices would help reduce the effect of these extreme values and make the distribution closer to normal, which can be useful for further statistical analysis.
